```python
# baseline

# 패키지 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## ID를 쓰지말자

```python
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
sub = pd.read_csv('sample_submission.csv')
```

```python
categorical_labels = ['Reason for absence', 'Month of absence', 'Day of the week', 'Seasons', 'Education']
continuous_labels = ['Transportation expense', 'Distance from Residence to Work', 'Service time', 'Age',
                     'Work load Average/day ', 'Hit target', 'Son', 'Pet', 'Weight', 'Height', 'Body mass index']
boolean_labels = ['Disciplinary failure', 'Social drinker', 'Social smoker']
```

```python
categorical = train.loc[:, categorical_labels]
categorical
```

```python
features = train.loc[:, 'Reason for absence':'Body mass index']
target = train['Absenteeism time in hours']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(features, target, test_size=0.15)
```

```python
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder()
OHE.fit(train[categorical_labels])
features_OHE = OHE.transform(features[categorical_labels])
pd.DataFrame(features_OHE.toarray())
```

```python
# ID를 고려 할 필요가 있을까? 굳이?
from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()
st_scaler.fit(features[continuous_labels])  # X_train에 대해서만 scaling하도록 해줘야 val이랑 test 조건이 맞는다!
pd.DataFrame(st_scaler.transform(features[continuous_labels]), columns=continuous_labels)
```

```python
def custom_preprocessing(train):
    categorical_features = pd.DataFrame(OHE.transform(train[categorical_labels]).toarray())
    continuous_features = pd.DataFrame(st_scaler.transform(train[continuous_labels]), columns=continuous_labels)
    bool_features = train[boolean_labels].reset_index(drop=True)

    return pd.concat([categorical_features, continuous_features, bool_features], axis=1)
```

```python
X_train_drop = X_train.drop(X_train[X_train['Reason for absence'] == 0].index).copy()
y_train_drop = y_train.drop(X_train[X_train['Reason for absence'] == 0].index).copy()
```

```python
X_train_drop.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
X_train_processed = custom_preprocessing(X_train_drop)
model = DecisionTreeRegressor(max_depth=5, random_state=9)
model.fit(X_train_processed, y_train_drop)
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_val_predict

param_grid = {"criterion": ["mse", "mae"],
              "min_samples_split": [10, 20, 40, 50, 60],
              "max_depth": [2, 5, 6, 8, 9, 10],
              "min_samples_leaf": [20, 40, 100],
              "max_leaf_nodes": [5, 20, 30, 50, 100],
              }

# Comment in order to publish in kaggle.

grid_cv_dtm = GridSearchCV(model, param_grid, cv=5)

grid_cv_dtm.fit(X_train_processed, y_train_drop)
```

```python
print("R-Squared::{}".format(grid_cv_dtm.best_score_))
print("Best Hyperparameters::\n{}".format(grid_cv_dtm.best_params_))
```

```python
mse_scores = cross_val_score(grid_cv_dtm.best_estimator_, X_train_processed,
                             y_train_drop, cv=10, scoring='neg_mean_squared_error')
```

```python

print("MSE::{:.3f}".format(np.mean(mse_scores)))
```

```python
best_dtm_model = grid_cv_dtm.best_estimator_
```

```python
test = custom_preprocessing(test)
```

```python
y_pred = best_dtm_model.predict(test)
```

```python
y_pred
```

```python
pd.DataFrame({'ID': range(0, len(y_pred)), 'Prediction': y_pred}).to_csv('submission.csv', index=False)
```

```python
#KNN 적용
from sklearn.model_selection import GridSearchCV
from sklearn import neighbors
params = {'n_neighbors':[2,3,4,5,6,7,8,9,10,15,20,25,30],'weights':['uniform', 'distance'],
          'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute']}
knn = neighbors.KNeighborsRegressor()
grid_KNN_model = GridSearchCV(knn, params, cv=5)
grid_KNN_model.fit(X_train_processed,y_train_drop)
```

```python
best_knn_model = grid_KNN_model.best_params_
best_knn_model
```

```python
best_knn_model = neighbors.KNeighborsRegressor(n_neighbors = 30,weights='uniform',algorithm='ball_tree')
best_knn_model.fit(X_train_processed,y_train_drop)
```

```python
y_pred = best_knn_model.predict(test)
pd.DataFrame({'ID': range(0, len(y_pred)), 'Prediction': y_pred}).to_csv('submission3.csv', index=False)
```
